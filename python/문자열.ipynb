```python
poem = '''all that doth flow we cannot liquid name
Or else would fire and water be the same;
but 'this the wet that makes it die,no doubt.' '''
word = 'the'
print(poem.find(word)) # 첫 문자 자리
print(poem.rfind(word)) #마지막 문자 자리
print(poem.index(word))
print(poem.rindex(word))
poem.capitalize() # 첫글자의 첫 알파뱃 대문자
poem.title() #첫 알파벳 대문자
poem.upper() #다 대문자, lower() 다 소문자, swapcase() 대소문자 바꾸기
```

```python
#!pip install matplotlib
import matplotlib.pyplot as plt
thee = '''How do I love thee? Let me count the ways.
I love thee to the depth and breadth and height
My soul can reach, when feeling out of sight
For the ends of Being and ideal Grace.
I love thee to the level of everyday's
Most quiet need, by sun and candle-light.
I love thee freely, as men strive for Right;
I love thee purely, as they turn from Praise.
I love thee with the passion put to use
In my old griefs, and with my childhood's faith.
I love thee with a love I seemed to lose
With my lost saint - I love thee with the breath,
Smiles, tears, of all my life! - and if God choose,
I shall but love thee better after death.'''
dist = thee.replace('?','')
dist = dist.replace('-','')
dist = dist.replace(',','')
dist = dist.replace('.','')
dist = dist.replace('!','')
dist1 = dist.split()
#dist2 = dist.split('\n')
dup = {}
for word in dist1:
    if word not in dup.keys():
        dup[word] = 1
    else:
        dup[word] += 1
print(dup)

x = [word for word in dup.keys()]
y = [word for word in dup.values()]
plt.figure(figsize=(100,15))
plt.bar(x,y, color=(0.2, 0.4, 0.6, 0.6),label='word')
plt.xlabel('Word',fontsize=30, color='w')
plt.ylabel('Count',fontsize=30,color='w')
plt.xticks(x, fontsize=22, color='r')
plt.title('Poem of Word Count',fontsize=50,color='r')
plt.rcParams['axes.grid']=True
plt.legend(loc=4)
plt.show()
# print(len(thee))
# word = '.'
# print(thee.find(word))
# print(thee.index(word))
# print(thee.count(word))
# print(thee[0:thee.find(word)]) #첫 sentence 프린트
```
